```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
train_dir = Path("../data/Train")
val_dir = Path("../data/val")

print("Train path:", train_dir)
print("Val path:", val_dir)
```

```python
classes = sorted([d.name for d in train_dir.iterdir() if d.is_dir()])

print("Jumlah kelas:", len(classes))
print(classes[:10], "...") 
```

```python
valid_ext = [".jpg", ".jpeg", ".png", ".jfif"]
```

```python
for c in classes:
    img_files = [
        p for p in (train_dir / c).iterdir()
        if p.suffix.lower() in valid_ext
    ]
    print(f"{c:<30}: {len(img_files)} images")
```

```python
#ambil gambar valid
sample_class = classes[0]
sample_dir = train_dir / sample_class
sample_images = [p for p in sample_dir.iterdir() if p.suffix.lower() in valid_ext]
sample_image_path = sample_images[0]
print("Sample image path:", sample_image_path)
#baca gambar
image = cv.imread(str(sample_image_path))
#ubah dari BGR ke RGB
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)    
#tampilkan gambar
plt.imshow(image)
plt.axis("off")
plt.show()
```
